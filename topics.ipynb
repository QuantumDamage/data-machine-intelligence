```python
data = pd.read_csv("Data. Machne. Intelligence. - Arkusz1.csv")
data = data.append(pd.read_csv("Data. Machne. Intelligence. - cross validated.csv"))
data = data.append(pd.read_csv("Data. Machne. Intelligence. - data science.csv"))
```

```python
data
```

```python
data["neighbours"] = data['neighbours'].str.split(', ')
```

```python
data = data.dropna(inplace=False)
```

```python
data
```

```python
list_of_edges = []
```

```python
for index, row in data.iterrows():
    result = list(zip(row["neighbours"], [row["topic"]]*len(row["neighbours"])))
    #print(result)
    #print(row["topic"]*len(row["neighbours"]))
    #break
    print(result)
    list_of_edges.append(result)
```
